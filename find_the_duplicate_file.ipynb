```python
import pandas as pd
import os

# CSV 파일 읽기
csv_file = "./data/data/bumper_data.csv"
df = pd.read_csv(csv_file)

# 모든 열을 기준으로 중복된 행 찾기 (특정 열을 기준으로 중복을 찾고 싶다면 subset 파라미터를 사용)
duplicates = df[df.duplicated(keep=False)]  # keep=False는 모든 중복을 True로 설정

# 중복된 행의 파일명 추출 (파일명이 첫 번째 열에 있다고 가정)
duplicate_filenames = duplicates.iloc[:, 0].tolist()


# 중복된 파일명 출력
print("중복된 파일명들:")
for filename in duplicate_filenames:
    file_name, file_extension = os.path.splitext(os.path.basename(filename))
    print(file_name)

# duplicates.to_csv("./duplicate_filenames.csv", index=False)
```
