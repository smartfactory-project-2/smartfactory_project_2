### 중복 파일 제거 및 추출

```python
import os
import csv
from collections import Counter


def find_and_delete_duplicates(csv_file, directory, column_name):
    # Step 1: CSV 파일에서 중복된 FullName 확인
    full_names = []
    with open(csv_file, mode="r", encoding="utf-8-sig") as file:
        reader = csv.DictReader(file)
        for row in reader:
            if column_name in row:
                full_names.append(row[column_name])

    # 중복 항목 찾기
    counter = Counter(full_names)
    duplicates = {name for name, count in counter.items() if count > 1}

    if not duplicates:
        print("중복된 파일이 없습니다.")
        return

    print("중복된 파일 목록:")
    for name in duplicates:
        print(name)

    # Step 2: JSON 파일 삭제
    for duplicate in duplicates:
        # JSON 파일명 생성
        json_file = f"{duplicate}.json"
        json_path = os.path.join(directory, json_file)
        if os.path.exists(json_path):
            os.remove(json_path)
            print(f"삭제: {json_path}")
        else:
            print(f"파일 없음: {json_path}")

    # Step 3: CSV 파일에서 중복된 항목 제거
    new_data = []
    with open(csv_file, mode="r", encoding="utf-8-sig") as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row[column_name] not in duplicates:
                new_data.append(row)

    # 새 CSV 파일 저장
    with open(csv_file, mode="w", newline="", encoding="utf-8-sig") as file:
        writer = csv.DictWriter(file, fieldnames=reader.fieldnames)
        writer.writeheader()
        writer.writerows(new_data)

    print("중복된 파일 및 관련 CSV 항목이 삭제되었습니다.")


# 경로 설정
csv_file = "output.csv"  # 생성된 CSV 파일 경로
directory = "/path/to/json/files"  # JSON 파일이 저장된 디렉토리 경로
column_name = "FullName"  # 중복 확인할 CSV 열 이름

# 함수 실행
find_and_delete_duplicates(csv_file, directory, column_name)
```
